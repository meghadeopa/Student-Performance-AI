## 1. Problem Statement
This project understands how a student's performance (test scores) is affected by variables such as:
- Gender
- Race/Ethnicity
- Parental Level of Education
- Lunch Type
- Test Preparation Course

## 2. Data Collection
- **Dataset Source:** https://www.kaggle.com/datasets/spscientist/students-performance-in-exams
- **Size:** 1000 rows × 8 columns

## 2.1 Import Required Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')

print('All libraries imported successfully!')
```

## 2.2 Load Dataset

```python
# Load the dataset
# Make sure stud.csv is in a 'data' folder in the same directory
df = pd.read_csv('data/stud.csv')
print(f'Dataset loaded successfully! Shape: {df.shape}')
df.head()
```

## 2.3 Dataset Column Information
- **gender** — sex of students (Male/Female)
- **race/ethnicity** — ethnicity (Group A, B, C, D, E)
- **parental level of education** — parent's final education level
- **lunch** — lunch type before test (standard / free-reduced)
- **test preparation course** — completed or not complete
- **math_score** — score in mathematics (0–100)
- **reading_score** — score in reading (17–100)
- **writing_score** — score in writing (10–100)

## 3.8 Feature Engineering — Adding Total Score and Average

```python
# Create Total Score and Average columns
df['total score'] = df['math_score'] + df['reading_score'] + df['writing_score']
df['average']     = df['total score'] / 3

print('New columns added: total score, average')
df.head()
```

```python
# Count full marks (100) per subject
math_full    = df[df['math_score']    == 100]['average'].count()
reading_full = df[df['reading_score'] == 100]['average'].count()
writing_full = df[df['writing_score'] == 100]['average'].count()

print(f'Full marks in Math:    {math_full}')
print(f'Full marks in Reading: {reading_full}')
print(f'Full marks in Writing: {writing_full}')
```

```python
# Count students scoring <= 20 per subject
math_less    = df[df['math_score']    <= 20]['average'].count()
reading_less = df[df['reading_score'] <= 20]['average'].count()
writing_less = df[df['writing_score'] <= 20]['average'].count()

print(f'Scored <=20 in Math:    {math_less}')
print(f'Scored <=20 in Reading: {reading_less}')
print(f'Scored <=20 in Writing: {writing_less}')
print('\nInsight: Students perform WORST in Mathematics')
```

## 4. Exploratory Data Analysis — Visualisations
### 4.1 Average Score Distribution (Histogram + KDE)

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 7))
plt.subplot(121)
sns.histplot(data=df, x='average', bins=30, kde=True, color='green')
plt.title('Distribution of Average Scores', fontsize=14)

plt.subplot(122)
sns.histplot(data=df, x='average', kde=True, hue='gender')
plt.title('Average Score by Gender', fontsize=14)
plt.show()
print('Insight: Female students tend to score higher overall')
```

### 4.2 Impact of Lunch Type on Performance

```python
plt.subplots(1, 3, figsize=(25, 6))
plt.subplot(141)
sns.histplot(data=df, x='average', kde=True, hue='lunch')
plt.title('Overall — Lunch Impact', fontsize=12)

plt.subplot(142)
sns.histplot(data=df[df.gender=='female'], x='average', kde=True, hue='lunch')
plt.title('Female — Lunch Impact', fontsize=12)

plt.subplot(143)
sns.histplot(data=df[df.gender=='male'], x='average', kde=True, hue='lunch')
plt.title('Male — Lunch Impact', fontsize=12)
plt.show()
print('Insight: Standard lunch consistently improves performance for both genders')
```

### 4.3 Impact of Parental Education

```python
plt.subplots(1, 3, figsize=(25, 6))
plt.subplot(141)
sns.histplot(data=df, x='average', kde=True, hue='parental level of education')
plt.title('Overall — Parental Education', fontsize=12)

plt.subplot(142)
sns.histplot(data=df[df.gender=='male'], x='average', kde=True, hue='parental level of education')
plt.title('Male — Parental Education', fontsize=12)

plt.subplot(143)
sns.histplot(data=df[df.gender=='female'], x='average', kde=True, hue='parental level of education')
plt.title('Female — Parental Education', fontsize=12)
plt.show()
print('Insight: Parental education helps male students more than female students')
```

### 4.4 Impact of Race/Ethnicity

```python
plt.subplots(1, 3, figsize=(25, 6))
plt.subplot(141)
sns.histplot(data=df, x='average', kde=True, hue='race/ethnicity')
plt.title('Overall — Race/Ethnicity', fontsize=12)

plt.subplot(142)
sns.histplot(data=df[df.gender=='female'], x='average', kde=True, hue='race/ethnicity')
plt.title('Female — Race/Ethnicity', fontsize=12)

plt.subplot(143)
sns.histplot(data=df[df.gender=='male'], x='average', kde=True, hue='race/ethnicity')
plt.title('Male — Race/Ethnicity', fontsize=12)
plt.show()
print('Insight: Group A and B score lowest. Group E scores highest regardless of gender.')
```

### 4.5 Violin Plots — Score Distribution per Subject

```python
plt.figure(figsize=(18, 8))

plt.subplot(1, 3, 1)
plt.title('MATH SCORES', fontsize=14)
sns.violinplot(y='math_score', data=df, color='red', linewidth=3)

plt.subplot(1, 3, 2)
plt.title('READING SCORES', fontsize=14)
sns.violinplot(y='reading_score', data=df, color='green', linewidth=3)

plt.subplot(1, 3, 3)
plt.title('WRITING SCORES', fontsize=14)
sns.violinplot(y='writing_score', data=df, color='blue', linewidth=3)

plt.suptitle('Score Distribution Across All Three Subjects', fontsize=16)
plt.tight_layout()
plt.show()
print('Insight: Most students score 60-80 in all subjects. Math has lowest tail.')
```

### 4.6 Multivariate Analysis — Pie Charts

```python
plt.rcParams['figure.figsize'] = (30, 12)

plt.subplot(1, 5, 1)
size = df['gender'].value_counts()
plt.pie(size, colors=['#ff6b6b','#4ecdc4'], labels=['Female','Male'], autopct='%.1f%%')
plt.title('Gender Distribution', fontsize=16)

plt.subplot(1, 5, 2)
size = df['race/ethnicity'].value_counts()
plt.pie(size, labels=size.index, autopct='%.1f%%')
plt.title('Race/Ethnicity Distribution', fontsize=16)

plt.subplot(1, 5, 3)
size = df['lunch'].value_counts()
plt.pie(size, colors=['#45b7d1','#f9ca24'], labels=size.index, autopct='%.1f%%')
plt.title('Lunch Type Distribution', fontsize=16)

plt.subplot(1, 5, 4)
size = df['test preparation course'].value_counts()
plt.pie(size, colors=['#6c5ce7','#fd79a8'], labels=size.index, autopct='%.1f%%')
plt.title('Test Prep Distribution', fontsize=16)

plt.subplot(1, 5, 5)
size = df['parental level of education'].value_counts()
plt.pie(size, labels=size.index, autopct='%.1f%%')
plt.title('Parental Education', fontsize=16)

plt.suptitle('Categorical Variable Distributions', fontsize=20, y=1.02)
plt.tight_layout()
plt.show()
```

### 4.7 Gender — Univariate and Bivariate Analysis

```python
# Gender count plot
f, ax = plt.subplots(1, 2, figsize=(18, 8))
sns.countplot(x=df['gender'], data=df, palette='bright', ax=ax[0], saturation=0.95)
for container in ax[0].containers:
    ax[0].bar_label(container, color='black', size=16)
ax[0].set_title('Gender Count', fontsize=14)

ax[1].pie(df['gender'].value_counts(), labels=['Female','Male'],
          explode=[0, 0.1], autopct='%1.1f%%', shadow=True,
          colors=['#ff4d4d','#4da6ff'])
ax[1].set_title('Gender Proportion', fontsize=14)
plt.show()
print('Insight: 518 Female (51.8%) vs 482 Male (48.2%) — nearly balanced dataset')
```

```python
# Gender vs scores
gender_group = df.groupby('gender')[['math_score','reading_score','writing_score','average']].mean()
print('Average scores by gender:')
print(gender_group.round(2))

gender_group[['math_score','reading_score','writing_score']].plot(
    kind='bar', figsize=(10, 6), colormap='Set2', rot=0, width=0.6)
plt.title('Average Score per Subject by Gender', fontsize=14)
plt.ylabel('Average Score')
plt.legend(['Math', 'Reading', 'Writing'])
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
print('Insight: Females outperform overall. Males score marginally higher in Math only.')
```

### 4.8 Race/Ethnicity Analysis

```python
# Race/Ethnicity count
f, ax = plt.subplots(1, 2, figsize=(18, 8))
sns.countplot(x=df['race/ethnicity'], data=df, palette='bright', ax=ax[0], saturation=0.95)
for container in ax[0].containers:
    ax[0].bar_label(container, color='black', size=14)
ax[0].set_title('Race/Ethnicity Count', fontsize=14)

ax[1].pie(df['race/ethnicity'].value_counts(),
          labels=df['race/ethnicity'].value_counts().index,
          explode=[0.1,0,0,0,0], autopct='%1.1f%%', shadow=True)
ax[1].set_title('Race/Ethnicity Proportion', fontsize=14)
plt.show()
print('Insight: Group C is largest. Group A is smallest.')
```

```python
# Race/Ethnicity vs scores
Group_data = df.groupby('race/ethnicity')
f, ax = plt.subplots(1, 3, figsize=(20, 8))

sns.barplot(x=Group_data['math_score'].mean().index,
            y=Group_data['math_score'].mean().values,
            palette='mako', ax=ax[0])
ax[0].set_title('Math Score by Group', fontsize=13)
for container in ax[0].containers:
    ax[0].bar_label(container, fmt='%.1f', color='black', size=11)

sns.barplot(x=Group_data['reading_score'].mean().index,
            y=Group_data['reading_score'].mean().values,
            palette='flare', ax=ax[1])
ax[1].set_title('Reading Score by Group', fontsize=13)
for container in ax[1].containers:
    ax[1].bar_label(container, fmt='%.1f', color='black', size=11)

sns.barplot(x=Group_data['writing_score'].mean().index,
            y=Group_data['writing_score'].mean().values,
            palette='crest', ax=ax[2])
ax[2].set_title('Writing Score by Group', fontsize=13)
for container in ax[2].containers:
    ax[2].bar_label(container, fmt='%.1f', color='black', size=11)

plt.suptitle('Average Scores by Race/Ethnicity Group', fontsize=16)
plt.tight_layout()
plt.show()
print('Insight: Group E highest scorer. Group A lowest scorer across all subjects.')
```

### 4.9 Parental Education Analysis

```python
# Parental education count
plt.figure(figsize=(12, 6))
sns.countplot(x=df['parental level of education'], data=df,
              palette='Blues_d', order=df['parental level of education'].value_counts().index)
plt.title('Parental Level of Education Distribution', fontsize=14)
plt.xticks(rotation=20, ha='right')
plt.tight_layout()
plt.show()
print('Insight: Most parents have some college education')
```

```python
# Parental education vs scores
df.groupby('parental level of education')[['math_score','reading_score','writing_score']].mean().plot(
    kind='barh', figsize=(12, 8), colormap='Set1')
plt.title('Average Scores by Parental Education Level', fontsize=14)
plt.xlabel('Average Score')
plt.legend(['Math', 'Reading', 'Writing'])
plt.tight_layout()
plt.show()
print('Insight: Students with master/bachelor degree parents score higher')
```

### 4.10 Lunch Type Analysis

```python
# Lunch count
plt.figure(figsize=(8, 5))
sns.countplot(x=df['lunch'], data=df, palette='PuBu')
plt.title('Lunch Type Distribution', fontsize=14)
plt.tight_layout()
plt.show()

# Lunch vs all 3 subjects with test prep
plt.figure(figsize=(14, 5))
plt.subplot(1, 3, 1)
sns.barplot(x=df['lunch'], y=df['math_score'], hue=df['test preparation course'])
plt.title('Lunch + Test Prep vs Math', fontsize=12)

plt.subplot(1, 3, 2)
sns.barplot(x=df['lunch'], y=df['reading_score'], hue=df['test preparation course'])
plt.title('Lunch + Test Prep vs Reading', fontsize=12)

plt.subplot(1, 3, 3)
sns.barplot(x=df['lunch'], y=df['writing_score'], hue=df['test preparation course'])
plt.title('Lunch + Test Prep vs Writing', fontsize=12)

plt.suptitle('Impact of Lunch Type and Test Preparation on All Subjects', fontsize=14)
plt.tight_layout()
plt.show()
print('Insight: Standard lunch + test prep = consistently highest scores across all subjects')
```

### 4.11 Outlier Detection — Boxplots

```python
plt.figure(figsize=(16, 5))

plt.subplot(1, 4, 1)
sns.boxplot(y=df['math_score'], color='skyblue')
plt.title('Math Score', fontsize=12)

plt.subplot(1, 4, 2)
sns.boxplot(y=df['reading_score'], color='hotpink')
plt.title('Reading Score', fontsize=12)

plt.subplot(1, 4, 3)
sns.boxplot(y=df['writing_score'], color='lightyellow')
plt.title('Writing Score', fontsize=12)

plt.subplot(1, 4, 4)
sns.boxplot(y=df['average'], color='lightgreen')
plt.title('Average Score', fontsize=12)

plt.suptitle('Outlier Detection — Boxplots', fontsize=14)
plt.tight_layout()
plt.show()
print('Insight: Math has low-end outliers. Reading and writing are more symmetric.')
```

### 4.12 Pairplot — Multivariate Analysis

```python
sns.pairplot(df[['math_score','reading_score','writing_score','average','gender']], hue='gender')
plt.suptitle('Pairplot — All Score Variables by Gender', y=1.02, fontsize=14)
plt.show()
print('Insight: All three scores increase linearly with each other — strong positive correlation')
```
